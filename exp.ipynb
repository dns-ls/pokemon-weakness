```python
def get_types(pokemon_en: str):
	row = a[a['name'] == pokemon_en]
	types = [row.type1, row.type2]
	types = [t.values[0] for t in types if pd.notna(t.values[0])]
	return types
```

```python
import pandas as pd

table = pd.read_csv("type_weaknesses.csv", index_col=0)



def get_type_effectiveness(type1: str, type2: str = "None"):
	if type2 == "None":
		return dict((table[type1]))
	else:
		return dict(table[type1] * table[type2])
```

```python
get_type_effectiveness("Fire", "Water")
```
